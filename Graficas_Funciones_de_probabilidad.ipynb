```python
import numpy as np
import matplotlib.pyplot as plt
```

# $x$

```python
# Coordenadas del triángulo
x = [0, 1, 1]  # Vertices en x
y = [0, 0, 1]  # Vertices en y

# Crear la figura
plt.figure(figsize=(6, 6))

# Dibujar las líneas que forman el triángulo
plt.plot([0, 1], [0, 1], color="indianred", label="y = x")
plt.axvline(1, color="cornflowerblue", linestyle='-', label="x = 1")

# Sombrear el área del triángulo
plt.fill(x, y, color="darkseagreen", alpha=0.5, hatch="//")

# Configurar los ejes
plt.axhline(0, color="black", linewidth=1)  # Eje x
plt.axvline(0, color="black", linewidth=1)  # Eje y
plt.xticks([0, 1], ["0", "1"], fontsize=14)
plt.yticks([0, 1], ["0", "1"], fontsize=14)

# Límites de la gráfica
plt.xlim(0., 1.1)
plt.ylim(0, 1.1)

# Etiquetas y título
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.title("$f(x)=x,x\in[0,1]$", fontsize=16)

# Mostrar la leyenda con tamaño de fuente ajustado
plt.legend(fontsize=14)

# Mostrar la gráfica
plt.grid(True, linestyle="--", linewidth=0.5)
plt.show()
```

# $f(x)=2x$

```python
# Definir la función f(x) = 2x
x = np.linspace(0, 1, 100)
y = 2 * x

# Crear la figura
plt.figure(figsize=(8, 6))

# Graficar la función
plt.plot(x, y, 'r-', linewidth=2, label=r"$f(x) = 2x$")

# Puntos específicos
x_points = [1/2, 3/4, 1]
y_points = [2 * x for x in x_points]

# Dibujar líneas punteadas desde el eje x hasta f(x)
for x_p, y_p in zip(x_points, y_points):
    plt.plot([x_p, x_p], [0, y_p], c='cornflowerblue', linestyle="--", linewidth=1)  # Línea vertical punteada

# Sombrear el área entre x=1/2 y x=3/4
x_shaded = np.linspace(1/2, 3/4, 100)
y_shaded = 2 * x_shaded
plt.fill_between(x_shaded, 0, y_shaded, color='darkseagreen', alpha=0.3, hatch='//')

# Etiquetas en los ejes
plt.xticks([0, 0.5, 0.75, 1], ['0', '1/2', '3/4', '1'])
plt.yticks([0, 1, 2], ['0', '1', '2'])
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Configurar el rango de los ejes
plt.xlim(0, 1.2)
plt.ylim(0, 2.2)

# Título y leyenda
plt.title("$f(x) = 2x,x\in[0,1]$", fontsize=16)
plt.legend(fontsize=12)

# Mostrar cuadrícula
plt.grid(alpha=0.3)

# Mostrar la gráfica
plt.show()
```

```python
# Crear los dominios de las diferentes partes de la función
x1 = np.linspace(0, 1, 100)  # Dominio para x^2
x2 = np.linspace(1, 2, 100)  # Dominio para y = 1

# Definir las respectivas funciones
y1 = x1**2
y2 = np.ones_like(x2)  # y = 1

# Crear la figura
plt.figure(figsize=(8, 6))

# Graficar cada parte de la función
plt.plot(x1, y1, c='cornflowerblue', label=r"$f(x) = x^2$")
plt.plot(x2, y2, c='indianred', label=r"$f(x) = 1$")

# Configurar los ejes
plt.xticks([0, 0.5, 1, 1.5, 2], ['0', '0.5', '1', '1.5', '2'])
plt.yticks([0, 0.5, 1, 1.5], ['0', '0.5', '1', '1.5'])
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Título y leyenda
plt.title("Gráfica de una función definida por partes", fontsize=16)
plt.legend(fontsize=12)

# Mostrar cuadrícula
plt.grid(alpha=0.3)

# Mostrar la gráfica
plt.show()
```

```python
# Definir valores de K
K = 1  # Cambia este valor según sea necesario

# Crear el dominio para x y y
x = np.linspace(0, 1, 50)  # Valores de x entre 0 y 1
y = np.linspace(0, 2, 50)  # Valores de y entre 0 y 2
x, y = np.meshgrid(x, y)   # Crear una malla para x e y

# Definir el plano z = K(x + 2y)
z = K * (x + 2 * y)

# Crear la figura en 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficar el plano
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8, edgecolor='k')

# Configurar límites para incluir el origen
ax.set_xlim(-1, 1.5)  # Ajustar límites en x
ax.set_ylim(-1, 2.5)  # Ajustar límites en y
ax.set_zlim(-1, 5)    # Ajustar límites en z

# Agregar líneas de los ejes en el origen
ax.plot([0, 0], [0, 0], [-1, 5], 'k--')  # Eje z
ax.plot([-1, 1.5], [0, 0], [0, 0], 'k--')  # Eje x
ax.plot([0, 0], [-1, 2.5], [0, 0], 'k--')  # Eje y

# Etiquetas de los ejes
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.set_title(f"Plano $z = K(x + 2y)$ con $K={K}$", fontsize=14)

# Mostrar la gráfica
plt.show()
```

```python
# Coordenadas del triángulo
x = [0, 3, 0]  # Vertices en x
y = [2, 0, 0]  # Vertices en y

# Crear la figura
plt.figure(figsize=(6, 4))

# Dibujar las líneas que forman el triángulo
plt.plot([3, 0], [0, 2], color="indianred", label=r"$y = -\frac{2}{3}x + 2$")

# Sombrear el área del triángulo
plt.fill(x, y, color="darkseagreen", alpha=0.5, hatch="//")

# Configurar los ejes
plt.axhline(0, color="black", linewidth=1)  # Eje x
plt.axvline(0, color="black", linewidth=1)  # Eje y
plt.xticks([0, 1, 2, 3], ["0", "1", "2", "3"], fontsize=14)
plt.yticks([0, 1, 2], ["0", "1", "2"], fontsize=14)

# Límites de la gráfica
plt.xlim(0., 3.5)
plt.ylim(0, 2.5)

# Etiquetas y título
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.title("$f(x)=x,x\in[0,1]$", fontsize=16)

# Mostrar la leyenda con tamaño de fuente ajustado
plt.legend(fontsize=14)

# Mostrar la gráfica
plt.grid(True, linestyle="--", linewidth=0.5)
plt.show()
```

# $P(x\leq2,y\geq1)$

```python
# Crear los datos para el triángulo
x = np.linspace(0, 3/2, 100)  # Valores de x entre 0 y 3/2
y1 = np.full_like(x, 1)  # Línea constante y = 1
y2 = 2 - (2/3) * x  # Línea oblicua y = -2/3 * x + 2

# Crear la figura
plt.figure(figsize=(6, 4))

# Graficar la línea principal
x_line = np.linspace(0, 3, 100)
y_line = 2 - (2/3) * x_line
plt.plot(x_line, y_line, color="black", label="y = -2/3x + 2")

# Sombrear únicamente el triángulo
plt.fill_between(x, y1, y2, where=(y2 >= y1), color="darkseagreen", alpha=0.5, hatch='//')

# Agregar la línea punteada en y = 1 con dominio 0 ≤ x ≤ 3/2
x_punteada = np.linspace(0, 3/2, 100)
plt.plot(x_punteada, np.full_like(x_punteada, 1), linestyle=":", color="cornflowerblue", label="y = 1")

# Agregar la línea punteada en x = 3/2 con dominio 0 < y < 1
y_punteada = np.linspace(0, 1, 100)
plt.plot(np.full_like(y_punteada, 3/2), y_punteada, linestyle=":", color="indianred", label="x = 3/2")

# Configurar los ejes
plt.axhline(0, color="black", linewidth=1)  # Eje x
plt.axvline(0, color="black", linewidth=1)  # Eje y
plt.xticks([0, 3/2, 3], ["0", "3/2", "3"], fontsize=14)
plt.yticks([0, 1, 2], ["0", "1", "2"], fontsize=14)

# Límites de la gráfica
plt.xlim(0, 3.5)
plt.ylim(0, 2.2)

# Etiquetas y título
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.title("$P(x\leq2,y\geq1)$", fontsize=16)

# Mostrar la leyenda con tamaño de fuente ajustado
plt.legend(fontsize=14)

# Mostrar la gráfica
plt.grid(True, linestyle="--", linewidth=0.5)
plt.show()
```
